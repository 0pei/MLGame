```python
import pickle
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import  classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedShuffleSplit


file = open("../log/1P11.pickle", "rb")
data = pickle.load(file)
file.close()
game_info =  data['ml_2P']['scene_info']
#gb=data['ml_1P']['scene_info'][0]
game_command = data['ml_1P']['command']
print(data)


for i in range(1, 16):
    path = "../log/1P" + str(i) + ".pickle"
    file = open(path, "rb")
    data = pickle.load(file)
    game_info = game_info + data['ml_1P']['scene_info']
    game_command = game_command + data['ml_1P']['command']
    file.close()

for i in range(1,12):
    path = "../log/2P" + str(i) + ".pickle"
    file = open(path, "rb")
    data = pickle.load(file)
    game_info = game_info + data['ml_1P']['scene_info']
    game_command = game_command + data['ml_1P']['command']
    file.close()

#print(gb)
#print(len(game_command))
```

```python
file = open("../log/2P1.pickle", "rb")
data = pickle.load(file)
file.close()
game_info2 =  data['ml_2P']['scene_info']
#gb=data['ml_1P']['scene_info'][0]
game_command2 = data['ml_2P']['command']
#print(game_command2)

for i in range(1, 16):
    path = "../log/1P" + str(i) + ".pickle"
    file = open(path, "rb")
    data = pickle.load(file)
    game_info2 = game_info + data['ml_2P']['scene_info']
    game_command2 = game_command + data['ml_2P']['command']
    file.close()

for i in range(1,12):
    path = "../log/2P" + str(i) + ".pickle"
    file = open(path, "rb")
    data = pickle.load(file)
    game_info2 = game_info + data['ml_2P']['scene_info']
    game_command2 = game_command + data['ml_2P']['command']
    file.close()

# print(len(game_command))
print(game_command2)
```

```python
g = game_info[1]
g2 = game_info2[1]
feature = np.array([g['ball'][0], g['ball'][1], g['platform_1P'][0],g['platform_2P'][0]])
feature2 = np.array([g2['ball'][0], g2['ball'][1], g2['platform_1P'][0],g2['platform_2P'][0]])
game_command[1] = 0
game_command2[1] = 0
# print(feature)
# print(g)
# print(game_command[1])

for i in range(2, len(game_info) - 1):
    g = game_info[i]
    feature = np.vstack((feature, [g['ball'][0], g['ball'][1], g['platform_1P'][0],g['platform_2P'][0]]))
    if game_command[i] == "NONE": game_command[i] = 0
    elif game_command[i] == "MOVE_LEFT": game_command[i] = 1
    else: game_command[i] = 2
        
for i in range(2, len(game_info2) - 1):
    g2 = game_info2[i]
    feature2 = np.vstack((feature2, [g2['ball'][0], g2['ball'][1], g2['platform_1P'][0],g2['platform_2P'][0]]))
    if game_command2[i] == "NONE": game_command2[i] = 0
    elif game_command2[i] == "MOVE_LEFT": game_command2[i] = 1
    else: game_command2[i] = 2
    
answer = np.array(game_command[1:-1])
answer2 = np.array(game_command2[1:-1])
# answer = np.append(answer, answer2)
# feature = np.vstack((feature, feature2))
# print(feature2)
# print(feature2.shape)
# print(feature.shape)
#print(answer)
#print(answer2)
# print(answer.shape)
```

```python
#資料劃分
x_train, x_test, y_train, y_test = train_test_split(feature, answer, test_size=0.33, random_state=42)
#參數區間
param_grid = {'n_neighbors':[1, 2, 3]}
# #交叉驗證 
cv = StratifiedShuffleSplit(n_splits=2, test_size=0.3, random_state=12)
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=cv, verbose=10, n_jobs=-1) #n_jobs為平行運算的數量
grid.fit(x_train, y_train)
grid_predictions = grid.predict(x_test)

# knn = KNeighborsClassifier(n_neighbors=1)
# knn.fit(x_train,y_train)
# pred = knn.predict(x_test)
# print(confusion_matrix(y_test,pred))
# print(classification_report(y_test,pred))

#儲存
file = open('model.pickle', 'wb')
pickle.dump(grid, file)
file.close()
```

```python
#資料劃分
x_train, x_test, y_train, y_test = train_test_split(feature2, answer2, test_size=0.33, random_state=42)
#參數區間
param_grid = {'n_neighbors':[1, 2, 3]}
# #交叉驗證 
cv = StratifiedShuffleSplit(n_splits=2, test_size=0.3, random_state=12)
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=cv, verbose=10, n_jobs=-1) #n_jobs為平行運算的數量
grid.fit(x_train, y_train)
grid_predictions = grid.predict(x_test)

#儲存
file = open('model2.pickle', 'wb')
pickle.dump(grid, file)
file.close()
```

```python
#最佳參數
print(grid.best_params_)
#預測結果
print(grid_predictions)
#混淆矩陣
print(confusion_matrix(y_test, grid_predictions))
#分類結果
print(classification_report(y_test, grid_predictions))
```
